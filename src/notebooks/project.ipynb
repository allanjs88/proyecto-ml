
**El conjunto de datos contiene 9357 instancias de respuestas promedio por hora de una matriz de 5 sensores químicos de óxido metálico, integrados en un Dispositivo Multisensor de Calidad del Aire.** El dispositivo estaba ubicado en el campo, en una zona con alta contaminación a nivel de la carretera, dentro de una ciudad italiana. Los datos se registraron desde marzo de 2004 hasta febrero de 2005 (un año), representando uno de los registros más largos disponibles públicamente de dispositivos de sensores químicos de calidad del aire desplegados en el campo.

Las concentraciones promedio por hora de CO, hidrocarburos no metánicos, benceno, óxidos de nitrógeno totales (NOx) y dióxido de nitrógeno (NO2) se obtuvieron mediante un analizador certificado de referencia co-localizado. Existen evidencias de sensibilidades cruzadas, así como de desviaciones en los conceptos y los sensores, como se describe en De Vito et al., Sens. And Act. B, Vol. 129,2,2008 (se requiere cita), lo que eventualmente afecta las capacidades de los sensores para estimar concentraciones. Los valores faltantes se etiquetan con el valor -200.

**Este conjunto de datos se puede utilizar exclusivamente con fines de investigación. Los fines comerciales están totalmente excluidos.**

---

**Información de los atributos:**

0. **Fecha** (DD/MM/AAAA)  
1. **Hora** (HH.MM.SS)  
2. **Concentración verdadera promedio por hora de CO** en mg/m³ (analizador de referencia)  
3. **PT08.S1 (óxido de estaño):** respuesta promedio por hora del sensor (nominalmente dirigido al CO)  
4. **Concentración verdadera promedio por hora de hidrocarburos no metánicos (NMHC)** en µg/m³ (analizador de referencia)  
5. **Concentración verdadera promedio por hora de benceno** en µg/m³ (analizador de referencia)  
6. **PT08.S2 (titania):** respuesta promedio por hora del sensor (nominalmente dirigido a los NMHC)  
7. **Concentración verdadera promedio por hora de NOx** en ppb (analizador de referencia)  
8. **PT08.S3 (óxido de tungsteno):** respuesta promedio por hora del sensor (nominalmente dirigido al NOx)  
9. **Concentración verdadera promedio por hora de NO2** en µg/m³ (analizador de referencia)  
10. **PT08.S4 (óxido de tungsteno):** respuesta promedio por hora del sensor (nominalmente dirigido al NO2)  
11. **PT08.S5 (óxido de indio):** respuesta promedio por hora del sensor (nominalmente dirigido al O3)  
12. **Temperatura** en °C  
13. **Humedad relativa** (%)  
14. **Humedad absoluta** (AH)

```python
# count the number of 0 values for each column
def count_ceros (dataset, columns):
    num_cero_values = (dataset[columns] == 0).sum()
    return num_cero_values

num_cero_values = count_ceros (df,['Date', 'Time', 'CO(GT)', 'PT08.S1(CO)', 'NMHC(GT)', 'C6H6(GT)', 'PT08.S2(NMHC)', 'NOx(GT)', 'PT08.S3(NOx)', 'NO2(GT)', 'PT08.S4(NO2)', 'PT08.S5(O3)', 'T', 'RH', 'AH'])

# Display the results
print("===== Columnas con ceros =====")
print(num_cero_values)
```
